 A partir de la columna `borrower_genders` creamos una nueva columna `gender_group` que clasifica cada préstamo según el género del prestatario:

*   Female = grupo formado solo por mujeres
*   Male   = grupo formado solo por hombres
*   Mixto  = combinación de hombres y mujeres

```python
def classify_gender(value):
    if pd.isna(value):
        return None
    genders = [g.strip() for g in value.split(',')] # Separar los géneros por coma
    unique = set(genders)                           # Eliminar duplicados
    if unique == {'female'}:
        return 'Female'
    elif unique == {'male'}:
        return 'Male'
    else:
        return 'Mixto'                              # Mezcla de female y male
```

Aplicamos la función a cada fila de `borrower_genders` y guardamos el resultado en `gender_group`

```python
df_kiva['gender_group'] = df_kiva['borrower_genders'].apply(classify_gender)
```

```python

print(df_kiva['gender_group'].value_counts())
```

En el siguiente gráfico hacemos uso de la columna derivada `year`, comparando el número de préstamos y la tasa de financiación por año.

```python
# Solicitudes por año
year_requests = df_kiva['year'].value_counts().sort_index()

# Tasa de financiación
df_kiva['funding_ratio'] = df_kiva['funded_amount'] / df_kiva['loan_amount']
year_funding = df_kiva.groupby('year')['funding_ratio'].mean()
```

```python
fig, ax1 = plt.subplots(figsize=(6,5))

bar_color = "#8FB9D9"
line_color = "#6D5B97"

# Barras → solicitudes
bars = ax1.bar(year_requests.index,
               year_requests.values,
               width=0.6,
               alpha=0.6,
               color=bar_color,
               label='Solicitudes de préstamo')

ax1.set_ylabel('Número de solicitudes')
ax1.set_xlabel('Año')

# Línea → tasa de financiación
ax2 = ax1.twinx()
line, = ax2.plot(year_funding.index,
                 year_funding.values,
                 marker='o',
                 markersize=10,
                 linewidth=2.5,
                 color=line_color,
                 label='Tasa de financiación')

ax2.set_ylabel('Tasa de financiación')

# Escala coherente
ax2.set_ylim(0.9, 1.0)

# Formato porcentaje
ax2.yaxis.set_major_formatter(
    plt.FuncFormatter(lambda y, _: f'{y*100:.0f}%')
)

# Título
plt.title('Evolución de solicitudes y tasa de financiación')

# Ejes
plt.xticks(year_requests.index)

# Grid suave
ax1.grid(axis='y', linestyle='--', alpha=0.2)

# Estética limpia
ax1.spines['top'].set_visible(False)
ax2.spines['top'].set_visible(False)

# Leyenda arriba del gráfico
ax1.legend(handles=[bars, line],
           loc='upper center',
           bbox_to_anchor=(0.5, 1.20),
           ncol=2,
           frameon=False)

# Ajuste de espacio superior
plt.subplots_adjust(top=0.85)

plt.tight_layout()
plt.show()
```

A continuación, analizamos la distribución de los préstamos utilizando la variable `gender_group`.

Gráfico de barras que muestra la comparación de préstamos por país y
grupo de género en los 10 países con mayor número de operaciones.




```python
# Seleccionamos los 10 países con mayor número de préstamos, de la columna 'country'
# contamos cuantas veces aparaceze cada país y nos quedamos con los top 10. Y nos quedamos con el nombre del país.
top_countries = df_kiva['country'].value_counts().head(10).index
print(top_countries)
```

```python
# Creamos la figura del gráfico
plt.figure(figsize=(14, 6))

# Filtramos el DataFrame para incluir únicamente los países seleccionados.
# Agrupamos los datos por país y 'gender_group'.
# Calculamos el número de registros por cada combinación.
# Convertimos 'gender_group' en columnas mediante 'unstack()'
# para facilitar la visualización en barras agrupadas.
df_kiva[df_kiva['country'].isin(top_countries)] \
    .groupby(['country', 'gender_group']) \
    .size() \
    .unstack() \
    .plot(kind='bar', ax=plt.gca())

# Personalizamos el gráfico
plt.title('Comparación de préstamos por país y género en los 10 principales mercados')
plt.xlabel('País')
plt.ylabel('Número de préstamos')
# Rotamos las etiquetas del eje X para mejorar la legibilidad
plt.xticks(rotation=45, ha='right')
# Añadimos la leyenda
plt.legend(title='Gender Group')
# Ajustamos automáticamente los márgenes para evitar solapamientos
plt.tight_layout()
# Mostramos el gráfico
plt.show()
```
