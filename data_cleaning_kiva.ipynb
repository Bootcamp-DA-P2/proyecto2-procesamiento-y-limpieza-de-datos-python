### Normalización de texto

```python
# Se analizan todas las columnas de tipo object (texto)
# para detectar posibles problemas de formato relacionados con espacios.

df_kiva.select_dtypes(include='object').apply(

    # Para cada columna se evalúan tres posibles problemas:
    lambda col: col.str.contains(
        r'^\s|\s$|\s{2,}',  # 1. espacios al inicio (^\\s)
                            # 2. espacios al final (\\s$)
                            # 3. múltiples espacios internos (\\s{2,})
        na=False
    ).sum()

).sort_values(ascending=False)
```

```python
# Aplicación de limpieza de espacios en todas las columnas de texto
for col in df_kiva.select_dtypes(include='object').columns:
    df_kiva[col] = (
        df_kiva[col]
        .str.strip()
        .str.replace(r'\s+', ' ', regex=True)
    )
```

```python
# Se realiza una verificación posterior al proceso de limpieza
# para confirmar que se han eliminado los problemas de formato

df_kiva.select_dtypes(include='object').apply(
    lambda col: col.str.contains(
        r'^\s|\s$|\s{2,}',
        na=False
    ).sum()

).sort_values(ascending=False)
```

Se han analizado todas las columnas de tipo texto para detectar problemas de formato relacionados con espacios en blanco. Aunque únicamente se han encontrado incidencias en `use` y `region`, la limpieza se ha aplicado a todas las variables de tipo object para garantizar la consistencia del dataset.

El proceso ha consistido en eliminar espacios iniciales y finales con `strip()` y normalizar los espacios múltiples. Tras una verificación posterior, se confirma que ya no existen problemas de este tipo en ninguna columna textual.

### Creación de variable categórica

 A partir de la columna `borrower_genders` creamos una nueva columna `gender_group` que clasifica cada préstamo según el género del prestatario:

*   Female = grupo formado solo por mujeres
*   Male   = grupo formado solo por hombres
*   Mixto  = combinación de hombres y mujeres

```python
def classify_gender(value):
    if pd.isna(value):
        return None
    genders = [g.strip() for g in value.split(',')] # Separar los géneros por coma
    unique = set(genders)                           # Eliminar duplicados
    if unique == {'female'}:
        return 'Female'
    elif unique == {'male'}:
        return 'Male'
    else:
        return 'Mixto'                              # Mezcla de female y male
```

Aplicamos la función a cada fila de `borrower_genders` y guardamos el resultado en `gender_group`.

```python
df_kiva['gender_group'] = df_kiva['borrower_genders'].apply(classify_gender)
```

Conteo de registros por categoría para verificar la distribución.

```python
print(df_kiva['gender_group'].value_counts())
```

Comparamos las columnas original y nueva para validar la clasificación, mostrando las 5 filas primeras.

En el segundo gráfico hemos decidido mostrar los 5 sectores mas populares y en que cantidad de paises lo son.

```python
# 1. Agrupamos por sector, contamos países únicos y tomamos los 5 más altos
top_5_sectores = (df_kiva.groupby('sector')['country']
                  .nunique()
                  .sort_values(ascending=False)
                  .head(5)
                  .reset_index())

# 2. Creamos el gráfico (Corrigiendo el FutureWarning)
plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_5_sectores,
    x='country',
    y='sector',
    hue='sector',   # Asignamos hue para evitar el aviso
    palette='magma',
    legend=False    # Quitamos la leyenda ya que el eje Y ya indica el sector
)

# 3. Añadimos etiquetas de datos para saber el número exacto
for index, value in enumerate(top_5_sectores['country']):
    plt.text(value, index, f' {value} países', va='center', fontsize=11, fontweight='bold')

# 4. Estética final
plt.title('Top 5 Sectores con mayor presencia internacional', fontsize=14, pad=15)
plt.xlabel('Cantidad de Países Únicos')
plt.ylabel('Sector')
plt.tight_layout()

plt.show()
```

El gráfico anterior muestra el Top 5 de sectores con mayor presencia internacional, medido a través de la métrica de Cantidad de Países en los que predominan esos sectores.

El gráfico demuestra que la globalización se distribuye en diferentes sectores. No hay una brecha significativa que indique el dominio absoluto de un solo sector sobre los demás.

En el siguiente gráfico hacemos uso de la columna derivada `year`, comparando el número de préstamos y la tasa de financiación por año.

```python
# Solicitudes por año
year_requests = df_kiva['year'].value_counts().sort_index()

# Tasa de financiación
df_kiva['funding_ratio'] = df_kiva['funded_amount'] / df_kiva['loan_amount']
year_funding = df_kiva.groupby('year')['funding_ratio'].mean()
```

```python
fig, ax1 = plt.subplots(figsize=(8,5))

# Barras: solicitudes
bars = ax1.bar(year_requests.index, year_requests.values,
               width=0.6, alpha=0.6, label='Solicitudes de préstamo')

ax1.set_xlabel('Año')
ax1.set_ylabel('Número de solicitudes')
ax1.set_xticks(year_requests.index)

# Línea: tasa de financiación
ax2 = ax1.twinx()
line, = ax2.plot(year_funding.index, year_funding.values,
                 marker='o', linewidth=2.5,
                 label='Tasa de financiación')

ax2.set_ylabel('Tasa de financiación')
ax2.set_ylim(0.9, 1.0)
ax2.yaxis.set_major_formatter(lambda y, _: f'{y*100:.0f}%')

# Sin cuadrícula
ax1.grid(False)
ax2.grid(False)

# Título
plt.title('Evolución de solicitudes y tasa de financiación')

# Leyenda
ax1.legend([bars, line],
           ['Solicitudes de préstamo', 'Tasa de financiación'],
           loc='upper center',
           bbox_to_anchor=(0.5, 1.20),
           ncol=2,
           frameon=False)

# Eliminación de bordes
ax1.spines['top'].set_visible(False)
ax2.spines['top'].set_visible(False)

plt.tight_layout()
plt.show()
```

A continuación, analizamos la distribución de los préstamos utilizando la variable `gender_group`.

Gráfico de barras que muestra la comparación de préstamos por país y
grupo de género en los 10 países con mayor número de operaciones.




```python
# Seleccionamos los 10 países con mayor número de préstamos, de la columna 'country'
# contamos cuantas veces aparaceze cada país y nos quedamos con los top 10. Y nos quedamos con el nombre del país.
top_countries = df_kiva['country'].value_counts().head(10).index
print(top_countries)
```

```python
# Creamos la figura del gráfico
plt.figure(figsize=(14, 6))

# Filtramos el DataFrame para incluir únicamente los países seleccionados.
# Agrupamos los datos por país y 'gender_group'.
# Calculamos el número de registros por cada combinación.
# Convertimos 'gender_group' en columnas mediante 'unstack()'
# para facilitar la visualización en barras agrupadas.
df_kiva[df_kiva['country'].isin(top_countries)] \
    .groupby(['country', 'gender_group']) \
    .size() \
    .unstack() \
    .plot(kind='bar', ax=plt.gca())

# Personalizamos el gráfico
plt.title('Comparación de préstamos por país y género en los 10 principales mercados')
plt.xlabel('País')
plt.ylabel('Número de préstamos')
# Rotamos las etiquetas del eje X para mejorar la legibilidad
plt.xticks(rotation=45, ha='right')
# Añadimos la leyenda
plt.legend(title='Gender Group')
# Ajustamos automáticamente los márgenes para evitar solapamientos
plt.tight_layout()
# Mostramos el gráfico
plt.show()
```

Y gráfico circular que muestra la distribución de préstamos según el género en Filipinas.

```python
# Filtramos el dataset para obtener únicamente los registros de Filipinas
# y calculamos la frecuencia de cada categoría de 'gender_group'
filipinas = df_kiva[df_kiva['country'] == 'Philippines']['gender_group'].value_counts()
print(filipinas)
```

```python
# Creamos un gráfico circular para visualizar la distribución
plt.figure(figsize=(7, 7))
plt.pie(
    filipinas,
    # Generamos etiquetas personalizadas que incluyen:
    # - nombre de la categoría
    # - valor absoluto formateado con separador de miles
    labels=[f'{label}\n({value:,})' for label, value in zip(filipinas.index, filipinas.values)],
    # Mostramos el porcentaje dentro de cada porción con un decimal
    autopct='%1.1f%%',
    # Definimos el ángulo inicial para comenzar desde la parte superior
    startangle=90
)
# Título del gráfico
plt.title('Distribución de préstamos según el género en Filipinas')
# Mostramos el gráfico
plt.show()
```
