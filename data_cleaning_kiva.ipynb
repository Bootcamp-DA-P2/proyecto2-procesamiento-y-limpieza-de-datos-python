```python
# Ver ejemplo de valores en las columnas tipo 'object'.
df_kiva.select_dtypes(include='object').head()
```

```python
# Se analizan todas las columnas de tipo object (texto)
# para detectar posibles problemas de formato relacionados con espacios.

df_kiva.select_dtypes(include='object').apply(

    # Para cada columna se evalúan tres posibles problemas:
    lambda col: col.str.contains(
        r'^\s|\s$|\s{2,}',  # 1. espacios al inicio (^\\s)
                            # 2. espacios al final (\\s$)
                            # 3. múltiples espacios internos (\\s{2,})
        na=False
    ).sum()

).sort_values(ascending=False)
```

```python
# Aplicación de limpieza de espacios en todas las columnas de texto
for col in df_kiva.select_dtypes(include='object').columns:
    df_kiva[col] = (
        df_kiva[col]
        .str.strip()
        .str.replace(r'\s+', ' ', regex=True)
    )
```

```python
# Se realiza una verificación posterior al proceso de limpieza
# para confirmar que se han eliminado los problemas de formato

df_kiva.select_dtypes(include='object').apply(
    lambda col: col.str.contains(
        r'^\s|\s$|\s{2,}',
        na=False
    ).sum()

).sort_values(ascending=False)
```

Se realiza una validación previa y posterior al proceso de limpieza de las columnas de tipo texto del dataset. En primer lugar, se detectan posibles inconsistencias relacionadas con espacios en blanco, incluyendo espacios al inicio, al final y múltiples espacios consecutivos dentro de los textos. Posteriormente, se aplica un proceso de limpieza global utilizando `strip()` y una expresión regular para normalizar los espacios. Finalmente, se repite la validación inicial para confirmar la efectividad de la transformación, verificando la reducción de incidencias detectadas.
